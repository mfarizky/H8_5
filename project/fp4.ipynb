# Preprocessing

# Define Model

```python
# Specify the number of components (number of dimensions after PCA)
n_components = 2

# Apply PCA
pca = PCA(n_components=n_components)
principal_components = pca.fit_transform(df)
```

```python
data_pca=pd.DataFrame(data=principal_components, columns=['pca1','pca2'])
data_pca.head()
```

```python
# Plot the explained variance for each principal component
plt.bar(range(1, n_components + 1), pca.explained_variance_ratio_)
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.show()
```

```python
# Choose the number of components based on the explained variance
# For example, if 95% of the variance is explained
n_components = len(pca.explained_variance_ratio_[pca.explained_variance_ratio_.cumsum() <= 0.95])
```

```python
pca = PCA(n_components=n_components)
principal_components = pca.fit_transform(df)
```

```python
k = 3

kmeans = KMeans(n_clusters=k)
clusters = kmeans.fit_predict(principal_components)
```

```python
# Visualize the clusters in 2D space
plt.scatter(principal_components[:, 0], principal_components[:, 1], c=clusters, cmap='viridis')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA Clustering')
plt.show()
```

## DBSCAN (Density-Based Spatial Clustering of Applications with Noise):

```python
# Tentukan parameter DBSCAN (eps: radius tetangga, min_samples: jumlah tetangga minimum)
eps_value = 0.5
min_samples_value = 5

# Inisialisasi dan fit DBSCAN model
dbscan = DBSCAN(eps=eps_value, min_samples=min_samples_value)
clusters = dbscan.fit_predict(df)
```

```python
# Visualisasi hasil clustering
plt.scatter(df.iloc[:, 0], df.iloc[:, 1], c=clusters, cmap='viridis')
plt.title('DBSCAN Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
