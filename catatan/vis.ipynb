```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Data asal
data = np.array([[2018, 2019, 2020, 2021, 2022],
                 [1.65, 2.10, -2.54, 1.95, 2.16]])

# Mengubah array numpy menjadi DataFrame
df = pd.DataFrame(data.transpose(), columns=['Tahun', 'Nilai'])

# Mengubah tipe data tahun menjadi integer
df['Tahun'] = df['Tahun'].astype(int)

# Menampilkan DataFrame setelah perubahan
print(df)

# Visualisasi data
plt.figure(figsize=(8, 6))
plt.plot(df['Tahun'], df['Nilai'], marker='o', color='b', linestyle='-', linewidth=2, markersize=8)
plt.xlabel('Tahun')
plt.ylabel('Nilai')
plt.title('RATA-RATA NILAI PBV')
plt.grid(True)

plt.xticks(np.arange(2018,2022+1))
plt.yticks(np.arange(min(df['Nilai']), max(df['Nilai']) + 0.5, 0.5))

plt.show()
```

```python
# Data asal
data = np.array([[2018, 2019, 2020, 2021, 2022],
                 [1.65, 2.10, -2.54, 1.95, 2.16]])

# Mengubah array numpy menjadi DataFrame
df = pd.DataFrame(data.transpose(), columns=['Tahun', 'Nilai'])

# Mengubah tipe data tahun menjadi integer
df['Tahun'] = df['Tahun'].astype(int)

# Menampilkan DataFrame setelah perubahan
print(df)

# Visualisasi data
plt.figure(figsize=(8, 6))
plt.plot(df['Tahun'], df['Nilai'], marker='o', color='b', linestyle='-', linewidth=2, markersize=5)
plt.xlabel('Tahun')
plt.ylabel('Nilai')
plt.title('RATA-RATA NILAI PBV')
plt.grid(axis='y')

plt.xticks(np.arange(2018,2022+1))
plt.yticks(np.arange(-3, 3 + 0.5, 0.5))

for i, txt in enumerate(df['Nilai']):
    plt.text(df['Tahun'][i], df['Nilai'][i], str(txt), ha='right', va='bottom')

plt.show()
```
