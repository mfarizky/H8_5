```python
import matplotlib.pyplot as plt

# Data
data = {
    'Lyft': {'3.23': 34.0, '3.36': 34.0, '3.74': 30.0},
    # Add UberX data as needed
}

# Convert distance keys to float for better plotting
for company in data.keys():
    data[company] = {float(distance): price for distance, price in data[company].items()}

# Plotting
for company in data.keys():
    distances = list(data[company].keys())
    prices = list(data[company].values())
    plt.scatter(distances, prices, label=company)

plt.xlabel('Distance')
plt.ylabel('Price')
plt.title('UberX and Lyft Prices')
plt.legend()
plt.show()
```
